```python
from forecaster import MoiraiForecaster
import numpy as np
import pandas as pd
from plotly import graph_objects as go

# Generate data
data = np.sin(np.linspace(0, 10, 200)) + np.random.randn(200) * 0.1

# Make predictions
model = MoiraiForecaster()
preds = model.predict(data, 200)

# Create DataFrames
df_past = pd.DataFrame({'index': np.arange(len(data)), 'value': data})
df_pred = pd.DataFrame({'index': np.arange(len(data), len(data) + len(preds)), 'value': preds})

# Plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_past['index'], y=df_past['value'], mode='lines', name='data'))
fig.add_trace(go.Scatter(x=df_pred['index'], y=df_pred['value'], mode='lines', name='preds'))
fig.show()
```

```python
from uni2ts.model.moirai2 import Moirai2Forecast, Moirai2Module
import numpy as np
data = (np.sin(np.linspace(0, 200, 1000)) + np.random.randn(1000) * 0.1).flatten()
model = Moirai2Forecast(
    module=Moirai2Module.from_pretrained("Salesforce/moirai-2.0-R-small"),
    prediction_length=100,
    context_length=1000,
    target_dim=1,
    feat_dynamic_real_dim=0,
    past_feat_dynamic_real_dim=0,
)
preds = model.predict(past_target=[data])[0][5]

from plotly import graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(100), y=data, mode='lines', name='data'))
fig.add_trace(go.Scatter(x=np.arange(100, 200), y=preds, mode='lines', name='preds'))
fig.show()
```
